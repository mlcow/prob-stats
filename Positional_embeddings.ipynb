```
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
def positional_encoding(pos, d):
  def encoding_at_i(pos, i, d):
    if i % 2 == 0:
      return math.sin (pos/10000**(2*i/d))
    else:
      return math.cos (pos/10000**(2*i/d))
  return np.asarray([encoding_at_i(pos, i, d) for i in range(d)])
```

```
plt.plot(positional_encoding(1, 100))
```

```
fig, axs = plt.subplots(4,3, figsize=[15,10])
for i in range(12):
  x = int(i / 3)
  y = i % 3
  axs[x, y].plot(positional_encoding(i, 100))
  axs[x, y].title.set_text('pos={}'.format(i))
```
